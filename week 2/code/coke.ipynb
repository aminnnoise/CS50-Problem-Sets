```python
amount = 50

while amount > 0:
    print(f"Amount Due: {amount}")
    coin = int(input("Insert Coin: "))

    if coin in [5,10,25]:
        amount -= coin

print(f"Change owed: {abs(amount)}")
```

```python
amount = 50

while amount > 0:
    print(f"Amount Due: {amount}")
    coin = int(input("Insert Coin: "))

    if coin in [5,10,25]:
        amount -= coin

print(f"Change owed: {-amount}")
```
