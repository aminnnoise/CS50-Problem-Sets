```python
user = input('file name: ').strip().lower()

if user.endswith('.gif') or user.endswith('.png'):
    print("image/" + user.split('.')[-1])
elif user.endswith('.jpg') or user.endswith('.jpeg'):
    print("image/jpeg")
elif user.endswith('.pdf') or user.endswith('.zip'):
    print("application/" + user.split('.')[-1])
elif user.endswith('.txt'):
    print("text/plain")
else:
    print("application/octet-stream")
```
