```python
num = input('enter your math question: ')


if '+' in num :
    print(float(num.split('+')[0]) + float(num.split('+')[-1]))
elif '-' in num :
    print(float(num.split('-')[0]) - float(num.split('-')[-1]))
elif '*' in num :
    print(float(num.split('*')[0]) * float(num.split('*')[-1]))
elif '/' in num :
    print(float(num.split('/')[0]) / float(num.split('/')[-1]))
```
