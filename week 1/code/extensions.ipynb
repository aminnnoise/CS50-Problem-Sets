```python
user = input('file name: ').strip().lower()

if user.endswith('.gif') or user.endswith('.png'):
    print("image/" + user.split('.')[-1])
elif user.endswith('.jpg') or user.endswith('.jpeg'):
    print("image/jpeg")
elif user.endswith('.pdf') or user.endswith('.zip'):
    print("application/" + user.split('.')[-1])
elif user.endswith('.txt'):
    print("text/plain")
else:
    print("application/octet-stream")
```

```python
user = input('file name: ').strip().lower()

image = ['.gif','.png']
app = ['.pdf','.zip']
jpeg = ['.jpg','jpeg']


for i in image:
    if (i in user):
        print("image/" + i.replace(".","").strip())
        

for i in app:
    if (i in user):
        print("application/" + i.replace(".","").strip())
        


for i in jpeg:
    if (i in user):
        print("image/jpeg")

if ('.txt' in user):
    print("text/plain")
else:
    print("application/octet-stream")
```
