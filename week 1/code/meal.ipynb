```python
def main():
    num = input("What time is it? ").strip()
    hour = convert(num)
    if 7 <= hour <= 8:
        print("Breakfast time")
    elif 12 <= hour <= 13:
        print("lunch time")
    elif 18 <= hour <= 19:
        print("dinner time")


def convert(num):
    parts = num.split(":")
    if len(parts) != 2:
        raise ValueError("Time must be in HH:MM format")

    h, m = parts
    m = float(m)       
    h = float(h)
    return h + m/60


if __name__ == "__main__":
    main()
```
