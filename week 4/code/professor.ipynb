```python
from random import randint

def get_level():

    while True:
        try:
            temp = int(input("Level: "))
        except ValueError:
            continue
        else:
            if temp not in [1,2,3]:
                continue
            else:
                return temp

def generate_integer(level):

    if level == 1:
        return int(randint(0,10))
    elif level == 2:
        return int(randint(10,99))
    else:
        return int(randint(100,999))


def main():

    points = 0
    level = get_level()

    for i in range(10):
        counter = 0
        fNumber = generate_integer(level)
        sNumber = generate_integer(level)
        answer = fNumber + sNumber

        while True:
            try:
                check = int(input("{0} + {1} = ".format(fNumber,sNumber)))
            except ValueError:
                counter += 1
                continue
            else: 
                if counter == 2:
                    print("{0} + {1} = {2}".format(fNumber,sNumber,answer))
                    break
                
                if answer != check:
                    print("EEE")
                    counter += 1
                    continue
                else:
                    points += 1
                    break
    print(f"Score: {points}")

if __name__ == "__main__":
    main()
```
