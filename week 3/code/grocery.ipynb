```python
item_list = {}
while True:
    try :
        item = input('Item : ').strip().upper()
        if item not in item_list:
            item_list[item] = 1
            continue
        else:
            item_list[item] += 1
    except EOFError:
        for key , value in sorted(item_list.items()):
            print(f"{value} {key}")
        break
        
        
```
