```python
months = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December"
]

while True:
    date = input("Date: ").strip()
    if "/" in date :
        try:
                m,d,y = date.split("/")
                # m = month / d = day / y = year
                m = int(m)
                d = int(d)
                y = int(y)

                if (d > 31 or d < 1):
                    continue
                if( m > 12 or m < 1):
                    continue
                if y < 999:
                    continue

                print(f"{y:04}-{m:02}-{d:02}")
                break

        except ValueError:
            pass
    else:
        try:
            m,d,y = date.split(" ")
            if ("," in d):
                d = d.replace(","," ")
            else :
                continue
            y = int(y)
            d = int(d)

            if d > 31 or d < 1:
                continue
            if y < 999:
                continue

            if m.title() in months:
                location = months.index(m.title())
                if (location == d):
                    location+=1

                print(f"{y}-{location:02}-{d:02}")
                break
            else:
                continue

        except:
            pass
```
