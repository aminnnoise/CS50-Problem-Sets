```python
while True:
    try :
        tank = input('enter the amount of gas in your gas tank: ')
        x,y = tank.split('/')
        x,y = int(x),int(y)
        if x > y or x<0 or 7<=0 :
            continue
    except ValueError:
        pass
    else:
        try :
            result = int(round((x/y)*100))
        except ZeroDivisionError:
            pass

        else :
            if  result <= 1 :
                print("E")
                break
            elif result >= 99 :
                print("F")
                break
            else:
                print(f"{result}%")
                break
```
