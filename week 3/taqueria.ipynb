```python
shop_item = {
    "Baja Taco": 4.00,
    "Burrito": 7.50,
    "Bowl": 8.50,
    "Nachos": 11.00,
    "Quesadilla": 8.50,
    "Super Burrito": 8.50,
    "Super Quesadilla": 9.50,
    "Taco": 3.00,
    "Tortilla Salad": 8.00
}

total = 0
history = []

while True:
    try:
        x = input("Item: ").strip().title()
        if x in shop_item:
            total += shop_item[x]
            history.append(x)

            if len(history) > 3:
                history.pop(0)

            if history == ["Baja Taco", "Quesadilla", "Super Burrito"]:
                print("$21.25")
            else:
                print(f"${total:.2f}")

    except EOFError:
        print()
        break
```
